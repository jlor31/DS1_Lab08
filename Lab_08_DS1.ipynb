```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as date
#import seaborn as sns
```

#### Tarea 1: Importar los datos

```python
# Definimos las variables con las direccion url de los conjuntos de datos

casos_confirmados = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename=time_series_covid19_confirmed_global.csv'

casos_muertos = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_deaths_global.csv&filename=time_series_covid19_deaths_global.csv'

casos_recuperados = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_recovered_global.csv&filename=time_series_covid19_recovered_global.csv'
```

```python
#como todos los url devuelven archivos .csv, podemos pasarlo a pandas directamente

Df_confirmados = pd.read_csv(casos_confirmados)
Df_muertos = pd.read_csv(casos_muertos)
Df_recuperados = pd.read_csv(casos_recuperados)
```

Como podemos ver, hay una fila por pais con 1 columna por fecha, lo cual no es deseable. SerÃ­a preferible que se nombren: 'region, pais, lat, lon'. AdemÃ¡s los nombres de las columnas Deben estandarizarse.


#### Tarea 2.1: Estandarizar el nombre de las columnas

```python
# Renombramos cada una de las columnas:

nombres_std = {'Province/State' : 'region', 'Country/Region' : 'pais', 'Lat' : 'lat', 'Long' : 'lon' } #los nombres deseados

Df_confirmados = Df_confirmados.rename( columns = nombres_std)

Df_muertos = Df_muertos.rename( columns = nombres_std)

Df_recuperados = Df_recuperados.rename( columns = nombres_std)

    
```

```python
Df_muertos.head()  #Listo !
```

#### Tarea 2.2: Transponer las fechas, de modo que solo haya una columna "fecha"

```python
#Empleamos el metodo "melt" de pandas visto en clase:


Df_confirmados = Df_confirmados.melt(id_vars = ['pais', 'region', 'lat', 'lon'], var_name = 'fecha', value_name = 'confirmados')

Df_muertos = Df_muertos.melt(id_vars = ['pais', 'region', 'lat', 'lon'], var_name = 'fecha', value_name = 'muertos')

Df_recuperados = Df_recuperados.melt(id_vars = ['pais', 'region', 'lat', 'lon'], var_name = 'fecha', value_name = 'recuperados')
```

Ahora, debemos pasar as fechas a un formato entendible por python. Es decir un objeto "datetime"

#### Tarea 2.3: Convertir las fechas a formato datetime

```python
Df_confirmados['fecha'] = pd.to_datetime(Df_confirmados['fecha'], format = '%m/%d/%y')
Df_muertos['fecha'] = pd.to_datetime(Df_muertos['fecha'], format = '%m/%d/%y')
Df_recuperados['fecha'] = pd.to_datetime(Df_recuperados['fecha'], format = '%m/%d/%y')
```

#### Tarea 2.4 : Agrupar los datos por pais, ordenar las fechas

```python
Df_confirmados = Df_confirmados.filter(['pais', 'region', 'fecha','confirmados', 'lat', 'lon'])  #movemos las columnas fecha y 'confirmados' hacia adelante'

Df_confirmados = Df_confirmados.sort_values(['pais', 'region', 'fecha',  'lat', 'lon'])  #Ordenamos/agrupamos poniendo en orden: pais, region, fecha

Df_confirmados = Df_confirmados.reset_index(drop = True)  #Reiniciamos el indice de las filas
```

```python
Df_muertos = Df_muertos.filter(['pais', 'region', 'fecha', 'muertos', 'lat', 'lon'])  #movemos las columnas fecha y 'muertos' hacia adelante'

Df_muertos = Df_muertos.sort_values(['pais', 'region', 'fecha',   'lat', 'lon'])  #Ordenamos/agrupamos poniendo en orden: pais, region, fecha

Df_muertos = Df_muertos.reset_index(drop = True)  #Reiniciamos el indice de las filas
```

```python
Df_recuperados = Df_recuperados.filter(['pais', 'region', 'fecha', 'recuperados', 'lat', 'lon'])  #movemos las columnas fecha y 'recuperados' hacia adelante'

Df_recuperados = Df_recuperados.sort_values(['pais', 'region', 'fecha',  'lat', 'lon'])  #Ordenamos/agrupamos poniendo en orden: pais, region, fecha

Df_recuperados = Df_recuperados.reset_index(drop = True)  #Reiniciamos el indice de las filas
```

#### Tarea 5.1 : Calcular casos nuevos por dia

```python
#Basta con restar la columna siguiente de la anterior en la columna "confirmados":

casos_nuevos_dia = [Df.iloc[i]['confirmados'] - Df.iloc[i-1]['confirmados'] for i in range(1,len(Df))]

casos_nuevos_dia.insert(0, 0)  #La primera fila no tiene valor, se lo ponemos.

casos_nuevos_dia = {'casos_nuevos_dia' : casos_nuevos_dia}
```

```python
Df['casos_nuevos_dia'] = pd.DataFrame(casos_nuevos_dia)
```

```python
Df.head(2000)
```

Todo parece bien a excepcion de ciertos valores negativos que se generaron debidoa  anomalias de conteo, o cuando el dataframe cambia de un pais a otro. Estos valores negativos no tiene significado alguno, y los podemos reemplazr por 0 ya que no hubo cambios en la cantidad de casos confirmados.

```python
#cambiamos ls negativos por 0s y reordenamos las columnas

Df.loc[Df["casos_nuevos_dia"] < 0, "casos_nuevos_dia"] = 0  
```

#### Tarea 5.2 : Calcular casos infectados

Ahora queremos encontrar el numero de casos cada dia que no son ni muertos ni recuperados. O sea los casos infectados. Una resta entre columnas deberÃ­a de ser suficiente.

```python
infectados = Df["confirmados"] - Df["muertos"] - Df["recuperados"]  #Hacemos el calculo

Df['infectados'] = infectados  #Creamos la nueva columna

Df =  Df.filter(['pais', 'region', 'fecha', 'confirmados',
                 'casos_nuevos_dia', 'infectados', 'recuperados', 'muertos','lat', 'lon'])  #Reordenar los datos
```
